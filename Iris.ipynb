Q2 (b) If the telco was to offer Phone as a standalone service, which type of contract would encourage customer retention?


```python

Customer_Info.Contract.value_counts()
```

```python
phoneservice = pd.crosstab(Customer_Info.Churn, [Customer_Info.Contract, Customer_Info.PhoneService,], normalize = "columns"). plot(kind = "barh")
```

According to the analysis, the customer who their preference contrat for a phone service subscription is  month-to-month while a two year contract would encourage customer retention in terms of not churning if  telco was to offer phone as a standalone service.
